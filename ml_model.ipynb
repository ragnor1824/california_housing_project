```python
# Encode the categorical data

converted_dataset_df = pd.get_dummies(final_dataset_df,columns=['ocean_proximity','longitude_category','latitude_category','value_category','median_age_category','total_rooms_category','total_bedrooms_category','population_category','households_category','median_income_category'])
converted_dataset_df.head()
```

```python
# Change true/false output to numeric

converted_dataset_df = converted_dataset_df * 1
converted_dataset_df.head()
```

```python
#Split processed data into features and target arrays
y = converted_dataset_df[["value_category_14500-59500", "value_category_59501-104500", "value_category_104501-149500", "value_category_149501-194500", "value_category_194501-239500", "value_category_239501-284500", "value_category_284501-329500", "value_category_329501-374500", "value_category_374501-419500", "value_category_419501-464500"]]
x = converted_dataset_df.drop(columns=["value_category_14500-59500", "value_category_59501-104500", "value_category_104501-149500", "value_category_149501-194500", "value_category_194501-239500", "value_category_239501-284500", "value_category_284501-329500", "value_category_329501-374500", "value_category_374501-419500", "value_category_419501-464500"])

#Split the processed data into training and testing datasets
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, stratify=y)
```

```python
#Create a StandardScaler instance
scaler = StandardScaler()

#Fit the StandardScaler
x_scaler = scaler.fit(x_train)

#Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
```

```python
# Create the model

nn = tf.keras.models.Sequential()

input_features = x.shape[1]

# First hidden layer
nn.add(tf.keras.layers.Dense(units=40, activation='relu',input_dim=input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=40, activation='relu'))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=40, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=10, activation='softmax'))

# Check model structure
nn.summary()
```

```python
# Compile the model
nn.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Train the model
train_model = nn.fit(x_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model
model_loss, model_accuracy = nn.evaluate(x_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
