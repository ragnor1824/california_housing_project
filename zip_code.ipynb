```python
from geopy.geocoders import Nominatim
import csv

# Initialize the Nominatim geocoder
geoLoc = Nominatim(user_agent="GetLoc")

# Open the CSV file for reading
with open('housing.csv', 'r') as csvfile:
    # Create a CSV reader
    csvreader = csv.DictReader(csvfile)

    # Iterate through each row in the CSV file
    for row in csvreader:
        # Extract latitude and longitude from the CSV row
        latitude = row['latitude']
        longitude = row['longitude']

        # Combine latitude and longitude as a string
        coordinates = f"{latitude},{longitude}"

        # Reverse geocode the coordinates
        locname = geoLoc.reverse(coordinates)

        # Print the location information for the coordinates
        print(f"Coordinates: {coordinates}")
        print(f"Location: {locname}\n")
```
